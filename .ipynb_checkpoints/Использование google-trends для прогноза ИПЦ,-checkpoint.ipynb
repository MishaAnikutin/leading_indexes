# Создание опережающих индикаторы на основе частоты поисковых запросов в Google

На сайте https://trends.google.ru/trends/ собрана статистика популярности поисковых запросов (и тем). Появилась гипотеза, что по ним можно строить опережающие индикаторы, как по новостному индексу. Популярность запроса нормируется к общем учислу запросов и заключена от 0 до 100. Данные есть с 2004 года, поэтому выбрал эту платформу, а не Вордстат от яндекса (https://wordstat.yandex.ru/)

Для составления индекса можно брать ключевые слова, которые больше всего влияют на индекс (можно получить из анализа новостей) и получить по ним данные популярности запроса

Вообще на сайте также публикуются самые популярные запросы сейчас, по их семантике можно что то предсказывать даже

## Для прогноза ИПЦ лучше всего подходили следующие запросы:
- Инфляция (запрос оказался сезонным, видимо из-за периодов отчетностей, поэтому привел к % к соотв. периоду предыдущего года)
- Курс Юаня (курс доллара показывал худшее качество модели, а между собой они сильно коррелируют)
- Дефолт
- COVID-19
- Российско-украинская война

Последние брались в сумме (называется `Причина кризиса*`). Это было сделано т.к. они равны нулю всюду, кроме периода их кризиса, а их сумма можно сказать это популярность запроса конкретного кризиса в стране)

__В датафрейме `факторы google-trends.csv` собрано множество поисковых запросов, я постарался из них отобрать лучшее__

```python
import pandas as pd
```

```python
google_trends = pd.read_csv('./данные google-trends /факторы google-trends.csv', index_col='Unnamed: 0')

google_trends.head(3)
```

# 1. Предобработка данных

Приведем все к типу `datetime[ns]` и обрежем номер дня

```python
from datetime import datetime
```

```python
google_trends['date'] = pd.to_datetime(google_trends['Месяц'], format='%Y-%m')
google_trends.drop(columns=['Месяц'], inplace=True)

google_trends.info()
```

```python
google_trends['COVID-19'].unique()
```

Удалим символ `<1`

```python
google_trends['COVID-19'] = google_trends['COVID-19'].str \
    .replace('<1', '0') \
    .astype(int) 
```

```python
google_trends['Российско-украинская война'] = google_trends['Российско-украинская война'].str \
    .replace('<1', '0') \
    .astype(int) 
```

Оставим данные только с 2015 года (чтобы переменные с лагом не отрезали размеры датафрейма, возьмем на 3 месяца с запасом)

```python
google_trends = google_trends[google_trends.date >= datetime(year=2013, month=6, day=1)]
```

```python
google_trends['Причина кризиса*'] = google_trends['COVID-19'] + google_trends['Российско-украинская война']

google_trends['Инфляция %'] = (google_trends['Инфляция'] - google_trends['Инфляция'].shift(12)) / google_trends['Инфляция'].shift(12)
# google_trends['Кризис %'] = (google_trends['Кризис'] - google_trends['Кризис'].shift(12)) / google_trends['Кризис'].shift(12)
```

```python
google_trends = google_trends.dropna()
```

# 4. Создадим переменные с лагами

```python
df = ipc.merge(google_trends, on='date')
```

```python
df = df[df.date >= datetime(year=2015, month=1, day=1)]
```

```python
df = df.drop(columns=['Кризис', 'Инфляция', 'Доля юаня в поиске', 'Курс рубля', 'COVID-19', 'Российско-украинская война'])
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
columns = df.drop(columns=['date']).columns

for i in range(1, 4):
    for column in columns:
        df[f'{column}_lag_{i}'] = df[column].shift(i)

df = df.dropna()
```

## 4.2 Модель

```python
from catboost import CatBoostRegressor
```

```python
test_split_date = datetime(year=2023, month=4, day=1)

def train_test_split(exog_var):
    X_train = df[df.date < test_split_date][exog_var].fillna(method='bfill')
    Y_train = df[df.date < test_split_date].goal.fillna(method='bfill')
    
    X_test = df[df.date >= test_split_date][exog_var]
    Y_test = df[df.date >= test_split_date].goal
    
    return X_train, X_test, Y_train, Y_test
```

Соберем все признаки, чтобы на обученной модели поняли, какие несут больший вклад

```python
exog_var_1 = ['goal_lag_1', 'goal_lag_2', 'goal_lag_3', 'Инфляция %_lag_3',
       'Инфляция %_lag_2', 'Экспорт_lag_1', 'Инфляция %_lag_1',
       'Экспорт_lag_2', 'курс юаня_lag_1', 'Дефолт_lag_1', 'курс юаня_lag_2',
       'Причина кризиса*_lag_3', 'Причина кризиса*_lag_2', 'Дефолт_lag_2',
       'курс юаня_lag_3', 'Экспорт_lag_3', 'Причина кризиса*_lag_1',
       'Дефолт_lag_3', 'Импорт_lag_1', 'Импорт_lag_2', 'Импорт_lag_3',
       'Торговля_lag_1', 'Торговля_lag_2', 'Торговля_lag_3',
       'Заработная плата_lag_3', 'Заработная плата_lag_2',
       'Заработная плата_lag_1']
```

```python
X_train_1, X_test_1, Y_train_1, Y_test_1 = train_test_split(exog_var_1)

grid = {
    'iterations':    [6, 8, 16], 
    'depth':         [3, 6, 8],
    'l2_leaf_reg':   [0.005, 0.01,  0.1,  0.5, 1],
    'learning_rate': [0.001, 0.01, 0.1, 0.5, 1]
}


res = CatBoostRegressor(verbose=False, loss_function='RMSE')\
    .grid_search(
        grid, X_train_1, Y_train_1, 
        calc_cv_statistics=True,
        search_by_train_test_split=True,
        refit=True, verbose=0
)


model_1 = CatBoostRegressor(**res['params'], verbose = False)
model_1.fit(X_train_1, Y_train_1)
```

```python
res['params']
```

## 4.3 Качество модели

```python
from sklearn.metrics import mean_absolute_percentage_error, r2_score, mean_absolute_error
```

```python
train_predict = model_1.predict(X_train_1)
test_predict = model_1.predict(X_test_1)

print(f'r2_score:   {np.round(r2_score(Y_train_1, train_predict), 3)}')
print(f'MAE train: {np.round(mean_absolute_percentage_error(Y_train_1, train_predict), 3) * 100:0.1f}%')
print(f'MAE test:  {np.round(mean_absolute_percentage_error(Y_test_1, test_predict), 3) * 100:0.1f}%')
```

Модель очевидно переобучилась из-за огромного числа параметров. При размере датасета ~100 записей

```python
fig, ax = plt.subplots(figsize=(12, 8))

ax.set_title('Индекс потребительских цен')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.plot(df[df.date < test_split_date].date, Y_train_1, label='Train')
ax.plot(df[df.date < test_split_date].date, train_predict, label='Train predict')

ax.plot(df[df.date >= test_split_date].date, Y_test_1, label='Test')
ax.plot(df[df.date >= test_split_date].date, test_predict, label='Test predict')

ax.grid(axis='x')
ax.legend()
```
